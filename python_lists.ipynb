#Q15.Find the largest number in a list manually.
##nums = [12, 5, 9, 34, 7]

```python
nums = [12, 5, 9, 34, 7]

largest = nums[0]

i = 1
while i < len(nums):
    if nums[i] > largest:
        largest = nums[i]
    i += 1

print(largest)
```

#Q16. Write a program using nested while loops to print a multiplication table (from 1 to 5).

```python
i = 1

while i <= 5:
    j = 1

    while j <= 5:
        print(i * j, end=" ")
        j += 1

    print()
    i += 1
```

#Q17. Use nested while loops to print the following pattern

```python
*
**
***
****
```

```python
i = 1

while i <= 4:
    j = 1

    while j <= i:
        print("*", end="")
        j += 1

    print()
    i += 1
```

##Write a program to reverse a list manually using a while loop.
##nums = [1, 2, 3, 4, 5]

```python
nums = [1, 2, 3, 4, 5]
reversed_list = []

i = len(nums) - 1

while i >= 0:
    reversed_list.append(nums[i])
    i -= 1

print(reversed_list)
```

#Q19.Find the Second Largest Element

##Write a program that finds the second largest number in a list using a while loop, without using sort() or max().
##nums = [10, 40, 20, 50, 30]

```python
nums = [10, 40, 20, 50, 30]

largest = nums[0]
second = nums[0]

i = 1
while i < len(nums):
    if nums[i] > largest:
        second = largest
        largest = nums[i]
    elif nums[i] > second and nums[i] != largest:
        second = nums[i]

    i += 1

print("Second Largest:", second)
```

#Q20.Remove Duplicates (Keep First Occurrence)

##Write a program that creates a new list containing each element from the original list only once, preserving the first occurrence order. Do not use set or the in keyword. Use while loops only.
##nums = [2, 3, 2, 1, 3, 4, 1]

```python
nums = [2, 3, 2, 1, 3, 4, 1]

result = []

i = 0
while i < len(nums):
    found = False

    j = 0
    while j < len(result):
        if nums[i] == result[j]:
            found = True
            break
        j += 1

    if not found:
        result.append(nums[i])

    i += 1

print(result)
```

#Q21. Separate Even and Odd Numbers

##Write a program that separates the numbers of a list into two new lists: one for even numbers and one for odd numbers, using a while loop only.
##nums = [1, 2, 3, 4, 5, 6, 7]


```python
nums = [1, 2, 3, 4, 5, 6, 7]

even = []
odd = []

i = 0

while i < len(nums):
    if nums[i] % 2 == 0:
        even.append(nums[i])
    else:
        odd.append(nums[i])

    i += 1

print("Even:", even)
print("Odd:", odd)
```

#Q22. Write a program using while that takes user input numbers until they enter 0, then prints the sum.

```python
total = 0

while True:
    num = int(input("Enter a number: "))

    if num == 0:
        break

    total += num

print("Sum =", total)
```

#َ23:Write a program that takes numbers from a list and prints the sum of only the positive numbers using a while loop.

```python
nums = [5, -3, 10, -2, 7, -1]

total = 0

i = 0

while i < len(nums):
    if nums[i] > 0:
        total += nums[i]

    i += 1

print(total)
```

#Q24. Count Vowels in a String
##Write a program that counts how many vowels (a, e, i, o, u) are in a given string using a while loop.

```python
text = input("Enter a string: ")

count = 0
i = 0

while i < len(text):
    if text[i].lower() in "aeiou":
        count += 1

    i += 1

print("Number of vowels:", count)
```
