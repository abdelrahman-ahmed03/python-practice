## Difference Between Tuple and List

- List is mutable, meaning its elements can be changed after creation.
- Tuple is immutable, meaning its elements cannot be changed after creation.
- Lists use square brackets [].
- Tuples use parentheses ().

### Example:
A tuple is preferred for storing fixed data such as coordinates because it should not be modified.

#Write a function that takes a tuple of numbers and returns:
##The sum

##The average

##The maximum value
##(Return them as a tuple).

````python
def tuple_info(t):
    total = sum(t)
    average = total / len(t)
    maximum = max(t)

    return total, average, maximum


nums = (10, 20, 30, 40)

print(tuple_info(nums))
````

# Function returning multiple tuples

##Write a function that accepts a tuple of numbers and returns two tuples:

##one containing all even numbers,

##one containing all odd numbers.

````python
def split_even_odd(t):
    even = ()
    odd = ()

    for num in t:
        if num % 2 == 0:
            even += (num,)
        else:
            odd += (num,)

    return even, odd


nums = (1, 2, 3, 4, 5, 6, 7, 8)

even, odd = split_even_odd(nums)

print("Even:", even)
print("Odd:", odd)
````

#Write a function that uses recursion to calculate the sum of all elements in a tuple.

````python
def recursive_sum(t):
    if len(t) == 0:
        return 0

    return t[0] + recursive_sum(t[1:])


nums = (10, 20, 30, 40)

print(recursive_sum(nums))
````

# Write a function that takes a tuple of numbers and returns the sum of all elements using a for loop.

````python
def tuple_sum(t):
    total = 0

    for num in t:
        total += num

    return total


nums = (5, 10, 15, 20)

print(tuple_sum(nums))
````

# Write a function that returns how many even numbers are in the tuple.

````python
def count_even(t):
    count = 0

    for num in t:
        if num % 2 == 0:
            count += 1

    return count


nums = (2, 5, 8, 11, 14, 17)

print(count_even(nums))
````

# Write a function that returns the smallest and largest numbers in a list

````python
def min_max(lst):
    return min(lst), max(lst)


numbers = [10, 25, 3, 99, 45]

smallest, largest = min_max(numbers)

print("Smallest:", smallest)
print("Largest:", largest)
````

# Write a function that returns the number of vowels (a, e, i, o, u) in a given string s.

````python
def count_vowels(s):
    count = 0

    for ch in s.lower():
        if ch in "aeiou":
            count += 1

    return count


text = input("Enter a string: ")

print(count_vowels(text))
````

# Ask the user to enter numbers repeatedly. Stop when the user enters 0, and then print the total sum of all the numbers entered.

````python
total = 0

while True:
    num = int(input("Enter a number: "))

    if num == 0:
        break

    total += num

print("Total =", total)
````

# Write a function that returns True if the word is a palindrome (reads the same backward) and False otherwise.

````python
def is_palindrome(word):
    return word == word[::-1]


word = input("Enter a word: ")

print(is_palindrome(word))
````

#Write a function that takes a tuple of numbers and returns a new tuple containing the square of each number using map().

````python
def square_tuple(t):
    return tuple(map(lambda x: x ** 2, t))


nums = (1, 2, 3, 4, 5)

print(square_tuple(nums))
````

#Write a function that takes a tuple of numbers and returns a new tuple containing only the prime numbers, using filter() and a lambda.

````python
def is_prime(n):
    if n < 2:
        return False

    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False

    return True


def prime_tuple(t):
    return tuple(filter(is_prime, t))


nums = (2, 3, 4, 5, 6, 7, 8, 9, 11, 13)

print(prime_tuple(nums))
````
