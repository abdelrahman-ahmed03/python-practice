#Q1.

#Write a program that keeps asking the user for a number until they enter 5, then print "You entered 5, exiting...".

```python
while True:
    num = int(input("Enter a number: "))

    if num == 5:
        print("You entered 5, exiting...")
        break
```

#Q4.

#Keep asking the user for numbers.

#If the number is even, print "Even number".

#If it is odd, print "Odd number".
#Stop when the user enters 0.

```python
while True:
    num = int(input("Enter a number: "))

    if num == 0:
        break

    if num % 2 == 0:
        print("Even number")
    else:
        print("Odd number")
```

#Q5.

#Ask the user for names. If a name has more than 5 characters, print it.
#Stop when the user types "stop".

```python
while True:
    name = input("Enter a name: ")

    if name.lower() == "stop":
        break

    if len(name) > 5:
        print(name)
```

#Q6.

#Ask the user for grades (0–100).

#If the grade ≥ 50, print "Pass".

#Else print "Fail".
#Stop if the user enters a number outside this range.

```python
while True:
    grade = int(input("Enter grade (0-100): "))

    if grade < 0 or grade > 100:
        break

    if grade >= 50:
        print("Pass")
    else:
        print("Fail")
```

#Q7.

#Keep asking the user for numbers.

#Print "Multiple of 3" if the number is divisible by 3.

#Print "Not multiple of 3" otherwise.
#Stop when the user enters -1.

```python
while True:
    num = int(input("Enter a number: "))

    if num == -1:
        break

    if num % 3 == 0:
        print("Multiple of 3")
    else:
        print("Not multiple of 3")
```

#Q8.

#Ask the user to enter words. Print "Palindrome!" if the word is a palindrome. Otherwise, print "Not palindrome". Stop when the user enters "end".

```python
while True:
    word = input("Enter a word: ")

    if word.lower() == "end":
        break

    if word == word[::-1]:
        print("Palindrome!")
    else:
        print("Not palindrome")
```

#Q9.

#Write a program that keeps asking the user for numbers and sums them up.

#If the sum becomes greater than 100, print "Limit reached" and stop.

```python
total = 0

while True:
    num = int(input("Enter a number: "))
    total += num

    if total > 100:
        print("Limit reached")
        break
```

#Q10.

#Create a number guessing game:

#The program asks the user to guess a secret number (say 42).

#If the guess is too low, print "Too low".

#If too high, print "Too high".

#Stop only when the guess is correct.

```python
secret = 42

while True:
    guess = int(input("Guess the number: "))

    if guess < secret:
        print("Too low")
    elif guess > secret:
        print("Too high")
    else:
        print("Correct!")
        break
```
