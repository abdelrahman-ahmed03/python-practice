## Q2) Robust calculator

Read two numbers and an operator from {+, -, *, /, //, %, **}. Compute the result with proper error handling (e.g., division by zero).

```python
num1 = float(input("Enter first number: "))
operator = input("Enter operator (+, -, *, /, //, %, **): ")
num2 = float(input("Enter second number: "))

if operator == "+":
    print(num1 + num2)

elif operator == "-":
    print(num1 - num2)

elif operator == "*":
    print(num1 * num2)

elif operator == "/":
    if num2 != 0:
        print(num1 / num2)
    else:
        print("Error: Division by zero")

elif operator == "//":
    if num2 != 0:
        print(num1 // num2)
    else:
        print("Error: Division by zero")

elif operator == "%":
    if num2 != 0:
        print(num1 % num2)
    else:
        print("Error: Division by zero")

elif operator == "**":
    print(num1 ** num2)

else:
    print("Invalid operator")
```

## Q3) Date validator (YYYY-MM-DD)

Read a date string 'YYYY-MM-DD'. Validate format and day/month ranges (assume Gregorian; implement leap-year rule). Print 'valid' or specific reason why invalid.

```python
date = input("Enter date (YYYY-MM-DD): ")

parts = date.split("-")

if len(parts) != 3:
    print("Invalid format")
else:
    year = int(parts[0])
    month = int(parts[1])
    day = int(parts[2])

    if month < 1 or month > 12:
        print("Invalid month")

    else:
        if month in [1, 3, 5, 7, 8, 10, 12]:
            max_day = 31

        elif month in [4, 6, 9, 11]:
            max_day = 30

        else:
            if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
                max_day = 29
            else:
                max_day = 28

        if day < 1 or day > max_day:
            print("Invalid day")
        else:
            print("Valid")
```

## Q4) Triangle classifier (+ right-angled)

Read three side lengths. First validate triangle inequality. If valid, classify as Equilateral/Isosceles/Scalene and also detect if Right-angled (Pythagoras), allowing small float tolerance.

```python
a = float(input("Side 1: "))
b = float(input("Side 2: "))
c = float(input("Side 3: "))

if a + b <= c or a + c <= b or b + c <= a:
    print("Invalid triangle")

else:
    if a == b == c:
        print("Equilateral")

    elif a == b or a == c or b == c:
        print("Isosceles")

    else:
        print("Scalene")

    sides = sorted([a, b, c])

    if abs(sides[0]**2 + sides[1]**2 - sides[2]**2) < 1e-6:
        print("Right-angled")
```

## Q6) Password policy with feedback

Read a password and validate: length ≥ 8, contains uppercase, lowercase, digit, and a special char (non-alphanumeric). Print True/False and list missing rules if any.

## Q12) Caesar cipher (letters only)

Read a string and an integer shift k. Shift alphabetic characters by k (wrapping A↔Z/a↔z), keep other chars untouched.

```python
text = input("Enter text: ")
k = int(input("Shift: "))

result = ""

for ch in text:
    if ch.isalpha():
        base = ord('A') if ch.isupper() else ord('a')
        result += chr((ord(ch) - base + k) % 26 + base)
    else:
        result += ch

print(result)
```

```python
text = input("Enter text: ")
k = int(input("Shift: "))

result = ""
i = 0

while i < len(text):
    ch = text[i]

    if ch.isalpha():
        base = ord('A') if ch.isupper() else ord('a')
        result += chr((ord(ch) - base + k) % 26 + base)
    else:
        result += ch

    i += 1

print(result)
```

## Q16) Second largest (distinct, no sort/max)

Given a list of integers, find the second largest **distinct** value without using `sort()` or `max()`.

```python
nums = [10, 40, 20, 50, 30, 50]

largest = None
second = None

for num in nums:
    if largest is None or num > largest:
        if num != largest:
            second = largest
            largest = num
    elif num != largest and (second is None or num > second):
        second = num

print(second)
```

## Q19) Merge two sorted lists (no sort)

Given two **already sorted** lists, merge them into a single sorted list without calling `sort()`.

```python
A = [1, 3, 5, 7]
B = [2, 4, 6, 8]

result = []
i = 0
j = 0

for _ in range(len(A) + len(B)):
    if i == len(A):
        result.append(B[j])
        j += 1
    elif j == len(B):
        result.append(A[i])
        i += 1
    elif A[i] <= B[j]:
        result.append(A[i])
        i += 1
    else:
        result.append(B[j])
        j += 1

print(result)
```

## Q20) Sliding window max sum

Given a list of integers and window size k, return the **maximum sum** among all contiguous windows of size k and the starting index of that window.

```python
nums = [2, 1, 5, 1, 3, 2]
k = 3

max_sum = sum(nums[:k])
current_sum = max_sum
start = 0

for i in range(k, len(nums)):
    current_sum = current_sum - nums[i - k] + nums[i]

    if current_sum > max_sum:
        max_sum = current_sum
        start = i - k + 1

print("Max Sum:", max_sum)
print("Start Index:", start)
```

## Q23) Remove items at Fibonacci positions

Remove items at 1‑based indices that are Fibonacci numbers (1, 2, 3, 5, 8, ... up to len(list)). Return the remaining list.

```python
nums = [10, 20, 30, 40, 50, 60, 70, 80]

fib = {1, 2}
a, b = 1, 2

while b <= len(nums):
    a, b = b, a + b
    fib.add(b)

result = []

for i in range(len(nums)):
    if (i + 1) not in fib:
        result.append(nums[i])

print(result)
```

## Q28) Collatz sequence stats

Read an integer n>0. Generate the Collatz sequence until it reaches 1. Print the sequence length and the maximum value seen.

```python
n = int(input("Enter number: "))

count = 1
maximum = n

while n != 1:
    if n % 2 == 0:
        n //= 2
    else:
        n = 3 * n + 1

    count += 1

    if n > maximum:
        maximum = n

print("Length:", count)
print("Maximum:", maximum)
```
