```python
nums = [1, 2, 2, 3, 1, 4, 5, 4]

result = []

for num in nums:
    if num not in result:
        result.append(num)

print(result)
```

#Q19) Reverse a string using a loop (without slicing).

```python
text = input("Enter a string: ")

reverse = ""

for ch in text:
    reverse = ch + reverse

print(reverse)
```

#Q20) Write a program that prints prime numbers between 1 and 20.

```python
for num in range(2, 21):
    prime = True

    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            prime = False
            break

    if prime:
        print(num, end=" ")
```

#Q21) Ask the user for 5 numbers, store them in a list, then print the average.

```python
numbers = []

for i in range(5):
    num = float(input("Enter a number: "))
    numbers.append(num)

average = sum(numbers) / len(numbers)

print("Average =", average)
```

#Q22) Find the longest word in a list.
##words = ["AI", "Machine", "Learning", "Python"]

```python
words = ["AI", "Machine", "Learning", "Python"]

longest = words[0]

for word in words:
    if len(word) > len(longest):
        longest = word

print(longest)
```

#Q24) Check if a string is a palindrome (the same forward and backward).

```python
text = input("Enter a string: ")

if text == text[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

#Q25) Write a program to find the factorial of a number using a while loop.

```python
num = int(input("Enter a number: "))

fact = 1

while num > 1:
    fact *= num
    num -= 1

print(fact)
```
